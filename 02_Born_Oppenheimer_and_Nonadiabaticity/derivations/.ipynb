```python
import sympy as sp

W11, W12, W22 = sp.symbols('W11, W12, W22')
theta = sp.symbols('theta')
```

```python
Cmat = sp.Matrix([[sp.cos(theta),sp.sin(theta)],
                  [-sp.sin(theta),sp.cos(theta) ]])

W = sp.Matrix([[W11,W12],
                  [W12,W22 ]])

Cinv = Cmat.transpose()

prod = W@Cmat
prod
```

```python
A = (Cinv@ prod)
A
```

```python
P, D = A.diagonalize()
```
